```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

'Загружаем файл и оставляем только интересующий нас промежуток времени'

event_df = pd.read_csv('7_4_Events.csv')
cond = (event_df['start_time'] >= '2018-01-01') & (event_df['start_time'] < '2019-01-01') & (event_df['event_type'] == 'registration')
registered = event_df[cond]['user_id'].to_list()
events = event_df[event_df['user_id'].isin(registered)]
events['start_time'] = pd.to_datetime(events['start_time'], format='%Y-%m-%dT%H:%M:%S')
display(events)
```

```python
'Пользователи совершившие оплату'

purchase_df = pd.read_csv('purchase.csv')
purchase_df['event_type'] = 'purchase'
purchase_df = purchase_df[purchase_df['user_id'].isin(registered)]
purchase_df['event_datetime'] = pd.to_datetime(purchase_df['event_datetime'], format='%Y-%m-%dT%H:%M:%S')
purchase_df
```

```python
'группа easy'

easy_level_df = events[events['selected_level'] == 'easy']
easy_list = easy_level_df['user_id'].to_list() 
purchase_easy = purchase_df[purchase_df.user_id.isin(easy_list)]
easy_users = easy_level_df['user_id'].nunique()
purchase_easy_users = purchase_easy['user_id'].nunique()
persent_easy_purchase = purchase_easy_users/easy_users
print ('процент оплативших от выбравших easy: {:.2%}'.format(persent_easy_purchase))

easy_reg = events[events.user_id.isin(easy_list)][events['event_type'] == 'registration']
merged_easy = easy_reg.merge(purchase_easy,on='user_id',how='inner')
merged_easy['deltatime'] = merged_easy['event_datetime']-merged_easy['start_time']
print ('среднее время от регистрации до оплаты:  ',   merged_easy['deltatime'].mean())

merged_easy1 = easy_level_df.merge(purchase_easy,on='user_id',how='inner')
merged_easy1['deltatime'] = merged_easy1['event_datetime']-merged_easy1['start_time']
print ('среднее время от выбора уровня до оплаты:', merged_easy1['deltatime'].mean())
```

```python
'группа medium'

medium_level_df = events[events['selected_level'] == 'medium']
medium_list =  medium_level_df['user_id'].to_list() # список пользователь выбравших средний уровень 
purchase_medium = purchase_df[purchase_df.user_id.isin(medium_list)]
medium_users = medium_level_df['user_id'].nunique()
purchase_medium_users = purchase_medium['user_id'].nunique()
persent_medium_purchase = purchase_medium_users/medium_users
print ('процент оплативших от выбравших medium: {:.2%}'.format(persent_medium_purchase))

medium_reg = events[events.user_id.isin(medium_list)][events['event_type'] == 'registration']
merged_medium = medium_reg.merge(purchase_medium,on='user_id',how='inner')
merged_medium['deltatime'] = merged_medium['event_datetime']-merged_medium['start_time']
print ('среднее время от регистрации до оплаты:   ', merged_medium['deltatime'].mean())

merged_medium1 = medium_level_df.merge(purchase_medium,on='user_id',how='inner')
merged_medium1['deltatime'] = merged_medium1['event_datetime']-merged_medium1['start_time']
print ('среднее время от выбора уровня до оплаты: ', merged_medium1['deltatime'].mean())
```

```python
'Группа hard'

hard_level_df = events[events['selected_level'] == 'hard']
hard_list =  hard_level_df['user_id'].to_list() # список пользователь выбравших средний уровень 
purchase_hard = purchase_df[purchase_df.user_id.isin(hard_list)]
hard_users = hard_level_df['user_id'].nunique()
purchase_hard_users = purchase_hard['user_id'].nunique()
persent_hard_purchase = purchase_hard_users/hard_users
print ('процент оплативших от выбравших hard: {:.2%}'.format(persent_hard_purchase))

hard_reg = events[events.user_id.isin(hard_list)][events['event_type'] == 'registration']
merged_hard = hard_reg.merge(purchase_hard,on='user_id',how='inner')
merged_hard['deltatime'] = merged_hard['event_datetime']-merged_hard['start_time']
print ('среднее время от регистрации до оплаты:  ', merged_hard['deltatime'].mean()) 

merged_hard1 = hard_level_df.merge(purchase_hard,on='user_id',how='inner')
merged_hard1['deltatime'] = merged_hard1['event_datetime']-merged_hard1['start_time']
print ('среднее время от выбора уровня до оплаты:', merged_hard1['deltatime'].mean())
```

Вывод: В ходе анализа можно выявить, что выбор уровня сложности и процент оплативших на прямую зависят друг от друга. 
       Чем ниже уровень, тем меньше процент отплат. 
       Так же, мы можем заметить, что нет существенной разницы во временных промежутках между регистрацией и оплаты,                  в зависимости от уровня сложности. Все временные интервалы находятся в диапазоне +-3 дня.
